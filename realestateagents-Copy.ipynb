```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
from time import sleep
import csv
import xlsxwriter
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as ec
import pandas as pd
```

```python
##############################
# opening to website
##############################
driver= webdriver.Chrome()
sleep(2)
url = 'https://www.top100realestateagents.com/listings/?search_keywords=&search_region=0&search_categories%5B%5D=&search_context=831#s=1'
driver.get(url)
sleep(10)
```

```python
page_source = BeautifulSoup(driver.page_source)
page_source
```

```python
# content-box

links = page_source.find_all("div",{"class":"content-box"})
#print(profiles)
sleep(2)


all_links = []
for link in links:
        all_links0 = []
        a = link.select("a")
        for i in a :
            link_URL = i.get('href')
            all_links0.append(link_URL)
        all_links. append(all_links0[0])   
            
#print(all_links)
print(len(all_links))
all_links
```

```python
##############################
# scrolling down the page to get all 
##############################
for j in range(0,100):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    sleep(2)
```
